```python
import numpy as np 
```

```python
#rand() function
var = np.random.rand(4)
print(var)
```

```python
var1 = np.random.rand(2, 5)
print(var1)
```

```python
#randn() function
var2 = np.random.randn(5)
print(var2)
```
