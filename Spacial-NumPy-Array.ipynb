```python
#Zeros
import numpy as np 
ar_zero = np.zeros(4)
ar_zero1 = np.zeros((3, 4))
print(ar_zero)
print()
print(ar_zero1)
```

```python
#Ones
ar_one = np.ones(4)
print(ar_one)
```

```python
#Empty
ar_em=np.empty(4)
print(ar_em)
```

```python
#Range
ar_rn = np.arange(4)
print(ar_rn)
```

```python
#Diagonal
ar_dia = np.eye(3)
print(ar_dia)
```

```python
ar_dia = np.eye(3, 5)
print(ar_dia)
```

```python
#Linspace
ar_lin = np.linspace(1, 20, num=5)
print(ar_lin)
```
